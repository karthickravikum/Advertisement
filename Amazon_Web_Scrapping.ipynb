```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
categories_urls = {
    "Laptops": "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops",
    "Tablets": "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets",
    "Phones": "https://webscraper.io/test-sites/e-commerce/allinone/phones/touch"
}
```

```python
titles, prices, ratings, descriptions, categories = [], [], [], [], []
```

```python
for category, url in categories_urls.items():
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    for product in soup.select('.thumbnail'):
        title = product.select_one('.title').text.strip()
        price = product.select_one('.price').text.strip()
        desc = product.select_one('.description').text.strip()


        rating_tag = product.select_one('[data-rating]')
        rating = int(rating_tag['data-rating']) if rating_tag and rating_tag.has_attr('data-rating') else 0

        titles.append(title)
        prices.append(price)
        descriptions.append(desc)
        ratings.append(rating)
        categories.append(category)
```

```python
df = pd.DataFrame({
    'Title': titles,
    'Category': categories,
    'Price': prices,
    'Rating': ratings,
    'Description': descriptions
})
```

```python
df.to_csv('/content/raw_data.csv', index=False)
df.to_json('/content/raw_data.json', orient='records', indent=4)

print(f"✅ Scraping complete! Total records: {len(df)}")
df.head(50)
```

```python
import pandas as pd

df = pd.read_csv('/content/raw_data.csv')
df.drop_duplicates(inplace=True)
df.fillna('Not Available', inplace=True)
df['Price'] = df['Price'].replace(r'[\$,]', '', regex=True).astype(float)
df['Title'] = df['Title'].str.title()
print("✅ Cleaned Data Sample:")
df.head()
df.to_csv('/content/cleaned_data.csv', index=False)
print("✅ Cleaned data saved as cleaned_data.csv")
```
